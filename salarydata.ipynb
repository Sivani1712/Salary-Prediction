```python
TITLE: SALARY PREDICTION
OBJECTIVE: To predict the salary of an employee using the no.of experience years
DATA SOURCE: https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv
```

```python
#regression problem
#STEP-1:import packages
import pandas as pd
```

```python
#STEP-2:read data
sal=pd.read_csv("https://github.com/YBI-Foundation/Dataset/raw/main/Salary%20Data.csv")
sal
```

```python
sal.head()
```

```python
sal.info()
```

```python
sal.describe()
```

```python
#STEP-3:define target(y) and features(x)
y=sal["Salary"]
x=sal[["Experience Years"]]
```

```python
y.shape
```

```python
x.shape
```

```python
#STEP-4:train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#STEP-5:select a model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#STEP-6:train the model
model.fit(x_train,y_train)
```

```python
c= model.intercept_
c
```

```python
m=model.coef_
m
```

```python
#STEP-7:prediction
y_pred=model.predict(x_test)
```

```python
x_test
```

```python
y_test
```

```python
#STEP-8:accuracy of the model
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```

```python
#for any value of Experience Years, the predicted salary is given by:
EY = int(input())
yp=c + m* EY
yp
```
